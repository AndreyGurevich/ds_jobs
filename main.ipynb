```python
import folium
import pandas as pd
import os
```

```python
DATA_FOLDER = "data"
```

```python
def generate_html_popup(company_name, url, comment):
    html = "<h1>"+company_name+"</h1><br>"+url
    return html

companies = pd.read_csv(os.path.join(DATA_FOLDER, "companies.csv"), sep=";", encoding='utf-8')
companies.fillna("", inplace = True)

companies
```

```python
map_osm = folium.Map(location=[55.7656, 37.6651], zoom_start=11)
for i in companies.iterrows():
    #print (i[1])
    folium.Marker([i[1].lat, i[1].lon],
                  popup=folium.Popup(folium.IFrame(html=generate_html_popup(i[1].company_name, i[1].url, i[1].comment)
                                                           , width=500, height=300), max_width=2650)
                 
                 ).add_to(map_osm)

map_osm
```
